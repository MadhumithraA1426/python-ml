```python
## linear regression
from sklearn.linear_model import LinearRegression
import numpy as np
X=np.array([[1],[2],[3],[4],[5]])
y=np.array([3,6,9,12,15])
model=LinearRegression()
model.fit(X,y)
print("Prediction for 6: ", model.predict([[6]]))
## Use: Continuous values (e.g., temperature prediction)
```

```python
## logistic regression
from sklearn.linear_model import LogisticRegression
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=LogisticRegression()
model.fit(X,y)
print("Prediction: ",model.predict([[2.5]]))
## Use: Yes/No, Failure/No Failure
```

```python
## decision tree
from sklearn.tree import DecisionTreeClassifier
import numpy as np
X=np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y=np.array([0,0,1,1,0,0,1,1])
model=DecisionTreeClassifier()
model.fit(X,y)
print("Prediction: ",model.predict([[3]]))
print("Prediction: ",model.predict([[6]]))
## Use: Rule-based decisions
```

```python
## random forest
from sklearn.ensemble import RandomForestClassifier
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=RandomForestClassifier()
model.fit(X,y)
print("Prediction:",model.predict([[2.5]]))
## Use: More accurate than decision tree
```

```python
## SVM
from sklearn.svm import SVC
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=SVC()
model.fit(X,y)
print("Prediction:",model.predict([[2.5]]))
## Use: Classification with clear boundary
```

```python
## KNN
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
print("Prediction:",model.predict([[2.5]]))
## Use: Based on nearest data points
```

```python
## Gradient boosting
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=GradientBoostingClassifier()
model.fit(X,y)
print("Prediction:",model.predict([[2.5]]))
## Use: High accuracy models (boosting)
```

```python
## Naive Bayes
from sklearn.naive_bayes import GaussianNB
import numpy as np
X=np.array([[1],[2],[3],[4]])
y=np.array([0,0,1,1])
model=GaussianNB()
model.fit(X,y)
print("Prediction:",model.predict([[2.5]]))
## Use: Probabilistic classification
```
